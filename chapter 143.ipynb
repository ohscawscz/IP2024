```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('hero.png')
res = cv2.resize(img,None,fx=50, fy=50, interpolation = cv2.INTER_CUBIC)
```

```python
img = cv2.imread('hero.png')
rows,cols = img.shape[:2]
M = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(img,M,(cols,rows))
```

```python
cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('hero.png')
rows,cols = img.shape[:2]
M = cv2.getRotationMatrix2D((cols/4,rows/4),45,1)
dst = cv2.warpAffine(img,M,(cols,rows))
```

```python
cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('drawing.png')
rows,cols,ch = img.shape

pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])

M = cv2.getAffineTransform(pts1,pts2)

dst = cv2.warpAffine(img,M,(cols,rows))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

```python
img = cv2.imread('dsu2.JPG')
rows,cols,ch = img.shape
pts1 = np.float32([[180,240],[816,216],[808,500],[180,500]])
pts2 = np.float32([[0,0],[800,0],[800,400],[0,400]])
#pts1에서 지정한 좌표들을 pts2로 불러오는 것
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(800,400))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
