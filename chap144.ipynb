```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#이미지를 컬러 모드로 불러옴 뒤에 0이 없어서
img = cv2.imread('ml.png')
kernel = np.ones((5,5),np.float32)/25 #5x5크기의 모든 값이 1인 커널을 생성한 후 각각의 값을 25로 나누어 평균 필터를 만듬 픽셀 주변의 값을 평균 내서 블러링을 만듬
dst = cv2.filter2D(img,-1,kernel) # 주변 픽셀값의 평균을 계산해 이미지의 블러 효과를 만듬
#blur3 = cv2.blur(img,(3,3))
blur5 = cv2.blur(img,(15,15))  # cv2.blur 함수를 사용하여 15x15 크기의 박스 필터를 적용하여 블러링 처리
#blur3 = cv2.GaussuanBlur(img,(5,5),0)
#GaussianBlur5 = cv2.GaussianBlur(img,(15,15),5)
MedianBlur5 = cv2.medianBlur(img, 15) # 이미지에 노이즈 제거를 수행 흐리게 만듬
BilateralFilter5 = cv2.bilateralFilter(img,9,75,75) #이미지 가장자리 정보를 보전하면서 블러링함 9=필터 크기 75= 색상 공간과 좌표 공간의 시그마 값
plt.subplot(121),plt.imshow(img),plt.title('blur5')
plt.xticks([]), plt.yticks([]) # 이 코드는 x축과 y축 눈금을 없애서 이미지만 표시되도록 함 
plt.subplot(122),plt.imshow(MedianBlur5),plt.title('MedianBlur5')
plt.xticks([]), plt.yticks([])
plt.show()
```
