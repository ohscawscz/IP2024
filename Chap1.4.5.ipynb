Chap 1.4.5 Morphological Transformations

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

##### Erosion

```python

img = cv2.imread('j.png',0) # 이미지를 0이니까 흑백으로 처리됨
kernel = np.ones((3,3),np.uint8) # 3x3 크기의 모든 값이 1인커널을 생성 커널은 erosion 연산에 사용
erosion = cv2.erode(img,kernel,iterations = 1) # 밝은 영역을 축소 시키는 효과가 있음 "1" 크기가 커지면 커질수록 output 이미지는 작아질거임
```

```python
plt.subplot(121),plt.imshow(img),plt.title('input') # subplot으로 이미지를 나란하게 보여줌 
plt.subplot(122),plt.imshow(erosion),plt.title('output') # erosion이 적용된 이미지를 표시
```

##### Dilation

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('jj.png',0)
kernel = np.ones((5,5),np.uint8)
dilation = cv2.dilate(img,kernel,iterations = 1)
```

```python
plt.subplot(121),plt.imshow(img),plt.title('input')
plt.subplot(122),plt.imshow(dilation),plt.title('output')
```

##### Opening

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('jjj.png',0)
kernel = np.ones((5,5),np.uint8)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
```

```python
plt.subplot(121),plt.imshow(img),plt.title('input')
plt.subplot(122),plt.imshow(opening),plt.title('output')
```
