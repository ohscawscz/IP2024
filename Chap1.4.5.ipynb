Chap 1.4.5 Morphological Transformations

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

##### Erosion

```python

img = cv2.imread('jj.png',0)
kernel = np.ones((3,3),np.uint8)
erosion = cv2.erode(img,kernel,iterations = 1)
```

```python
plt.subplot(121),plt.imshow(img),plt.title('input')
plt.subplot(122),plt.imshow(erosion),plt.title('output')
```

##### Dilation

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('jj.png',0)
kernel = np.ones((5,5),np.uint8)
dilation = cv2.dilate(img,kernel,iterations = 1)
```

```python
plt.subplot(121),plt.imshow(img),plt.title('input')
plt.subplot(122),plt.imshow(dilation),plt.title('output')
```

##### Opening

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('jjj.png',0)
kernel = np.ones((5,5),np.uint8)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
```

```python
plt.subplot(121),plt.imshow(img),plt.title('input')
plt.subplot(122),plt.imshow(opening),plt.title('output')
```
